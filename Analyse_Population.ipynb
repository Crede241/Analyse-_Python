## Visualisation des données :


```python
file_path = 'D:/Test transport/height_weight_sex_data.csv'
df_sex = pd.read_csv(file_path, sep=',')
df_sex
```

```python
sns.histplot(data=df_sex, x="Height")
```

```python
sns.boxplot(data=df_sex, x="Weight", hue="Sex")
```

```python
colors = {"Male": "blue", "Female": "orange"}

fig, ax = plt.subplots()
scatter = ax.scatter(df_sex['Height'],
                     df_sex['Weight'],
                     c=df_sex['Sex'].map(colors), alpha=0.7)
legend_handles = [mpatches.Patch(color=color, label=ind)
                  for ind, color in colors.items()]
ax.legend(handles=legend_handles, title="Sex")

ax.set_xlabel('Height')
ax.set_ylabel('Weight')
ax.set_title('Taille vs longueur par sex')
ax.grid(True)

plt.show()
```
